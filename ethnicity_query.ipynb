```python
import json
import numpy as np
import demjson
import pandas as pd
import urllib
from ipykernel import kernelapp as app
```

```python
# load data
data_path = "cleaned_us_AI-csv.csv"
unique_inventors = pd.read_csv(data_path)
query_eth_json_base = 'http://www.name-prism.com/api_eth/json/'
```

```python
# insert new colomns first
ethnicities = ['White', '2PRACE', 'Hispanic', 'Black', 'API', 'AIAN']
i = 10
for eth in ethnicities:
    new_col = pd.Series([0]*unique_inventors.shape[0])
    unique_inventors.insert(i, eth, new_col)
    i += 1
```

```python
cnt = 0
for i in range(unique_inventors.shape[0]):
    cnt += 1
    if cnt%100 == 0:
        print(cnt)
    f_name = str(unique_inventors.loc[i, 'inventor_name_first'])
    m_name = str(unique_inventors.loc[i, 'inventor_name_middle'])
    l_name = str(unique_inventors.loc[i, 'inventor_name_last'])
    """
    if not f_name.isalpha():
        f_name = f_name.split()[0]
        f_name = f_name.split('.')[0]
        if not f_name or not f_name.isalpha() or len(f_name) <= 1:
            unique_inventors.loc[i, 'valid'] = 'N'
            continue
    if not l_name.isalpha():
        l_name = l_name.split()[0]
        l_name = l_name.split('.')[0]
        if not l_name or not l_name.isalpha() or len(l_name) <= 1:
            unique_inventors.loc[i, 'valid'] = 'N'
            continue
    """
    # check the length of first name and middle name
    if len(f_name) < 2 and len(m_name) < 2:
        unique_inventors.loc[i, 'valid'] = 'N'
        continue
    if len(f_name) < 2:
        name = m_name + ' ' + l_name
    else:
        name = f_name + ' ' + l_name
    percent_code = urllib.request.quote(name)
    url = query_eth_json_base + percent_code
    
    try:
        response = urllib.request.urlopen(url).read()
    except:
        print(i, url)
    else:
        response = response.decode("utf-8") 
        result = json.loads(response)
        for eth in result:
            unique_inventors.loc[i, eth] = result[eth]
```
